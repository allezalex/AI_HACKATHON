```python
df['Direct emissions/ Scope 1 (metric tonnes CO2e) for Total generation of grid supplied energy'].fillna(df['Direct emissions (metric tonnes CO2e) for Total generation of grid-supplied energy'], inplace=True)
df['Direct emissions/ Scope 1 (metric tonnes CO2e) for Total emissions (excluding generation of grid-supplied energy)'].fillna(df['Direct emissions (metric tonnes CO2e) for Total emissions (excluding generation of grid-supplied energy)'], inplace=True)
df['Indirect emissions from use of grid supplied energy/Scope 2 (metric tonnes CO2e) for Total generation of grid supplied energy'].fillna(df['Indirect emissions from use of grid supplied energy (metric tonnes CO2e) for Total generation of grid supplied energy'], inplace=True)
df['Indirect emissions from use of grid supplied energy/Scope 2 (metric tonnes CO2e) for Total emissions (excluding generation of grid-supplied energy)'].fillna(df['Indirect emissions from use of grid supplied energy (metric tonnes CO2e) for Total Emissions (excluding generation of grid-supplied energy)'], inplace=True)
df['Emissions occurring outside city boundary/ Scope 3 (metric tonnes CO2e) for Total generation of grid supplied energy'].fillna(df['Emissions occurring outside city boundary (metric tonnes CO2e) for Total Generation of grid supplied energy'], inplace=True)
df['Emissions occurring outside city boundary/ Scope 3 (metric tonnes CO2e) for Total emissions (excluding generation of grid-supplied energy)'].fillna(df['Emissions occurring outside city boundary (metric tonnes CO2e) for Total Emissions (excluding generation of grid-supplied energy)'], inplace=True)
df['Total Scope 1 Emissions (metric tonnes CO2e)'].fillna(df['TOTAL Scope 1 Emissions (metric tonnes CO2e)'], inplace=True)
df['Total Scope 2 Emissions (metric tonnes CO2e)'].fillna(df['TOTAL Scope 2 emissions (metric tonnes CO2e)'], inplace=True)
df['Total Scope 3 emissions (metric tonnes CO2e)'].fillna(df['TOTAL Scope 3 Emissions'], inplace=True)
df['Total BASIC Emissions (GPC)'].fillna(df['TOTAL BASIC Emissions (GPC)'], inplace=True)
df['Total BASIC+ Emissions (GPC)'].fillna(df['TOTAL BASIC+ Emissions (GPC)'], inplace=True)
df['Change in emissions'].fillna(df['Chnage in emissions'], inplace=True)

cols_drop = ['Total emissions (metric tonnes CO2e)', 'Reason for increase/decrease in emissions ', 'Direct emissions (metric tonnes CO2e) for Total generation of grid-supplied energy',\
            'Direct emissions (metric tonnes CO2e) for Total emissions (excluding generation of grid-supplied energy)', 'Indirect emissions from use of grid supplied energy (metric tonnes CO2e) for Total generation of grid supplied energy',\
            'Indirect emissions from use of grid supplied energy (metric tonnes CO2e) for Total Emissions (excluding generation of grid-supplied energy)',\
            'Emissions occurring outside city boundary (metric tonnes CO2e) for Total Generation of grid supplied energy', 'Emissions occurring outside city boundary (metric tonnes CO2e) for Total Emissions (excluding generation of grid-supplied energy)',\
            'TOTAL Scope 1 Emissions (metric tonnes CO2e)', 'TOTAL Scope 2 emissions (metric tonnes CO2e)', 'TOTAL Scope 3 Emissions', 'TOTAL BASIC Emissions (GPC)', 'TOTAL BASIC+ Emissions (GPC)', 'Chnage in emissions']

df.drop(columns=cols_drop, inplace=True)
df.head()
```

```python
temp_need = temp.loc[temp['Year']>2015]
temp_need.drop(temp_need[temp_need['AvgTemperature']==-99].index, inplace=True)
temp_need['AvgTempCel'] = (temp_need['AvgTemperature']-30)/2
# df_final.drop(columns=['Average annual temperature (in Celsius)'], inplace=True)
for year in temp_need['Year'].unique():
    df_year = temp_need[temp_need['Year']==year]
    for city in df_year['City'].unique():
        df_city = df_year[df_year['City']==city]
        average_temp_year = sum(df_city['AvgTempCel'])/len(df_city)
        
        df_final.loc[(df_final['City']==city)&(df_final['Year Reported to CDP']==year),'Average Temperature'] = average_temp_year

df_final.head()
```
