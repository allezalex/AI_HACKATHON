```python
city_size = pd.read_csv('City_Size.csv', encoding='latin1')
city_size.head()
```

```python
for city in city_size['City / Urban area'].unique():
    city_ind = city_size[city_size['City / Urban area']==city]
    if '//' in city:
        city = city.split('//')
        if len(city)==2:
            df.loc[(df['City']==city[0])|(df['City']==city[1]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
        elif len(city)==3:
            df.loc[(df['City']==city[0])|(df['City']==city[1])|(df['City']==city[2]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
    elif ' / ' in city:
        city = city.split(' / ')
        if len(city)==2:
            df.loc[(df['City']==city[0])|(df['City']==city[1]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
        elif len(city)==3:
            df.loc[(df['City']==city[0])|(df['City']==city[1])|(df['City']==city[2]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
    elif '/' in city:
        city = city.split('/')
        if len(city)==2:
            df.loc[(df['City']==city[0])|(df['City']==city[1]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
        elif len(city)==3:
            df.loc[(df['City']==city[0])|(df['City']==city[1])|(df['City']==city[2]), 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
            continue
    else:
        df.loc[df['City']==city, 'Land Area (in km sq.)'] = city_ind['Land area (in sqKm)'].values[0]
        continue

df.head()
```

```python
df['Land Area (in km sq.)'].fillna(df['Land area (in square km)'], inplace=True)
df.drop(columns=['Land area (in square km)'], inplace=True)
```
